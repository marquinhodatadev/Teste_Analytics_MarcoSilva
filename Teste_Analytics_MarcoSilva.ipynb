# Parte 1 - Limpeza e Análise de Dados de Vendas

1.1 - Limpeza e Análise de Dados de Vendas:


```python
# Importando as bibliotecas necessárias:
import random  # Para gerar valores aleatórios.
from datetime import datetime, timedelta  # Para manipular datas.
import pandas as pd  # Para criar e manipular o dataset.
import numpy as np  # Para operações numéricas.
```

```python
# Criando listas para os produtos e suas categorias.
produtos = ["Mesa", "Cadeira", "Estante", "Sofá", "Guarda-Roupa"]  # Lista de móveis.
categorias = ["Sala", "Quarto", "Escritório"]  # Categorias dos móveis.

# Definindo a data inicial e final para o período de vendas.
data_inicial = datetime(2023, 1, 1)  # Começa em 01/01/2023
data_final = datetime(2023, 12, 31)  # Termina em 31/12/2023

# Criando uma lista para armazenar os dados simulados.
dados = []
```

```python
# Gerando os dados simulados.
for i in range(1, 51):  # Gera 50 registros.
    id_venda = i  # ID único.
    data_venda = data_inicial + timedelta(days=random.randint(0, (data_final - data_inicial).days))
    produto = random.choice(produtos)  # Escolhe um produto aleatório.
    categoria = random.choice(categorias)  # Escolhe uma categoria aleatória.
    quantidade = random.randint(1, 10)  # Quantidade entre 1 e 10.
    preco = round(random.uniform(50, 500), 2)  # Preço aleatório entre R$ 50 e R$ 500.

    # Adicionando os dados simulados à lista.
    dados.append([id_venda, data_venda.strftime("%d/%m/%Y"), produto, categoria, quantidade, preco])

print("Dados gerados com sucesso!")
```

```python
# Criando o DataFrame.
df_vendas = pd.DataFrame(dados, columns=["ID", "Data", "Produto", "Categoria", "Quantidade", "Preço"])

# Exibindo o DataFrame.
df_vendas
```

```python
# Verificando se há valores faltantes no DataFrame.
valores_faltantes = df_vendas.isnull().sum()  # Conta valores NaN por coluna.

# Exibindo o resultado.
if valores_faltantes.sum() > 0:
    print("Valores faltantes encontrados nas seguintes colunas:")
    print(valores_faltantes[valores_faltantes > 0])  # Exibe apenas as colunas com NaN.
else:
    print("Não há valores faltantes no DataFrame.")
```

```python
# Verificando se há duplicatas no DataFrame.
duplicatas = df_vendas.duplicated()  # Retorna uma Série indicando duplicatas.

# Exibindo o resultado.
if duplicatas.sum() > 0:
    print(f"Há {duplicatas.sum()} duplicatas no DataFrame.")
else:
    print("Não há duplicatas no DataFrame.")
```

```python
# Adicionando uma nova coluna para o total de vendas (Quantidade * Preço).
df_vendas['Total_Vendas'] = df_vendas['Quantidade'] * df_vendas['Preço']

# Calculando o total de vendas por produto.
total_vendas_produto = df_vendas.groupby('Produto')['Total_Vendas'].sum()

# Exibindo o resultado.
print("Total de vendas por produto:")
print(total_vendas_produto)
```

```python
# Encontrando o produto com o maior total de vendas.
produto_maior_venda = df_vendas.groupby('Produto')['Total_Vendas'].sum().idxmax()

# Exibindo o resultado.
print(f"O produto com o maior número de vendas totais é: {produto_maior_venda}")
```

1.2 - Análise Exploratória de Dados de Vendas

```python
import matplotlib.pyplot as plt

# Garantindo que a coluna 'Data' esteja no formato datetime.
df_vendas['Data'] = pd.to_datetime(df_vendas['Data'], errors='coerce')

# Criando uma coluna com o mês das vendas
df_vendas['Mes'] = df_vendas['Data'].dt.to_period('M')  # Extraindo mês e ano.

# Calculando as vendas totais por mês.
vendas_mensais = df_vendas.groupby('Mes')['Total_Vendas'].sum()

# Criando o gráfico de linha.
plt.figure(figsize=(10, 5))
plt.plot(vendas_mensais.index.astype(str), vendas_mensais.values, marker='o', linestyle='-', color='green')
plt.title('Tendência de Vendas Mensais', fontsize=14)
plt.xlabel('Mês', fontsize=12)
plt.ylabel('Total de Vendas', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 1. Total de vendas por mês.
vendas_por_mes = df_vendas.groupby('Mes')['Total_Vendas'].sum()

# 2. Produto mais vendido em quantidade.
produto_mais_vendido = df_vendas.groupby('Produto')['Quantidade'].sum().idxmax()
quantidade_mais_vendido = df_vendas.groupby('Produto')['Quantidade'].sum().max()

# 3. Categoria com maior receita.
categoria_maior_receita = df_vendas.groupby('Categoria')['Total_Vendas'].sum().idxmax()
receita_maior_categoria = df_vendas.groupby('Categoria')['Total_Vendas'].sum().max()

# 4. Identificando o mês com maior venda total.
mes_maior_venda = vendas_por_mes.idxmax()
total_maior_venda = vendas_por_mes.max()

# 5. Média de vendas por transação.
media_vendas_transacao = df_vendas['Total_Vendas'].mean()

# Exibindo os insights.
print("Padrões e Insights Observados:")
print(f"1. O total de vendas por mês varia significativamente. O mês com maior venda foi {mes_maior_venda}, com um total de R$ {total_maior_venda:.2f}.")
print(f"2. O produto mais vendido foi '{produto_mais_vendido}' com {quantidade_mais_vendido} unidades vendidas.")
print(f"3. A categoria com maior receita foi '{categoria_maior_receita}', gerando R$ {receita_maior_categoria:.2f}.")
print(f"4. A média de vendas por transação foi de R$ {media_vendas_transacao:.2f}.")
print("5. As vendas apresentam uma sazonalidade visível, com picos em determinados meses do ano.")
```

# Parte 2: SQL

```python
# Abrindo ou criando o arquivo para salvar as consultas SQL.
arquivo_sql = open('consultas_sql.sql', 'w')
```

```python
# Consulta 1: Listar o nome do produto, categoria e o total de vendas (Quantidade * Preço).
consulta_1 = """
-- Listar o nome do produto, categoria e o total de vendas (Quantidade * Preço)
SELECT
    Produto,
    Categoria,
    SUM(Quantidade * Preço) AS Total_Vendas
FROM vendas
GROUP BY Produto, Categoria
ORDER BY Total_Vendas DESC;
"""

# Primeira consulta.
arquivo_sql.write("-- Consulta 1: Total de vendas por produto e categoria\n")
arquivo_sql.write(consulta_1 + "\n\n")
```

```python
# Consulta 2: Identificar os produtos que venderam menos no mês de junho de 2024.
consulta_2 = """
-- Identificar os produtos que venderam menos no mês de junho de 2024
SELECT
    Produto,
    SUM(Quantidade) AS Total_Quantidade
FROM vendas
WHERE strftime('%Y-%m', Data) = '2024-06'
GROUP BY Produto
ORDER BY Total_Quantidade ASC;
"""

# Consulta 2.
arquivo_sql.write("-- Consulta 2: Produtos com menor venda em junho de 2024\n")
arquivo_sql.write(consulta_2 + "\n")
```

```python
# Salvando as consutas.
arquivo_sql.close()

print("Consultas SQL geradas e salvas no arquivo 'consultas_sql.sql'.")
```
